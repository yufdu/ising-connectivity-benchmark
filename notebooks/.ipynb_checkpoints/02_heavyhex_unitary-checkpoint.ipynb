# Notebook 02 — Heavy-Hex Unitary (SWAP Routing)

Transpile ideal Trotter circuits to heavy-hex topology.  The transpiler
inserts SWAPs to route non-local interactions.  This is the standard
approach and shows the **raw connectivity tax**.

```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))

import numpy as np
from qiskit.transpiler import CouplingMap
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

from src.model import coupling_pairs, exact_trotter_magnetization
from src.circuits import build_ideal_trotter_circuit, transpile_to_heavyhex
from src.metrics import extract_metrics, save_metrics, load_metrics
```

```python
N_QUBITS = 6
ALPHAS = [0.5, 1.0, 2.0, 4.0, 8.0]
CUTOFF_FRAC = 0.01  # keep pairs with J >= 1% of J_max
H_FIELD = 1.0
THETA = np.pi / 4
MAX_STEPS = 8
SEED = 42
GEOMETRY = "honeycomb"
```

## 1. Prepare heavy-hex coupling map

We use `from_heavy_hex(5)` — a 57-qubit heavy-hex lattice, well above
our 6-qubit needs.  This gives the transpiler ample room to route.

```python
heavy_cmap = CouplingMap.from_heavy_hex(5, bidirectional=True)
print(f"Heavy-hex: {heavy_cmap.size()} qubits, "
      f"{len(heavy_cmap.get_edges()) // 2} undirected edges")
```

## 2. Transpile and extract metrics

```python
from src.simulate import simulate_mz

SIMULATE_MZ = False # shortcut for metric extraction, see notebook 05 for simulation results

all_metrics = []

for alpha in ALPHAS:
    for n_steps in range(MAX_STEPS + 1):
        circ = build_ideal_trotter_circuit(
            N_QUBITS, alpha, THETA, n_steps, measure=True, cutoff_frac=CUTOFF_FRAC, geometry=GEOMETRY, 
        )
        circ_t = transpile_to_heavyhex(circ, heavy_cmap, seed=SEED)
        m = extract_metrics(circ_t, "unitary", N_QUBITS, alpha, n_steps, THETA, geometry=GEOMETRY)
        m.mz_exact = exact_trotter_magnetization(
            N_QUBITS, alpha, H_FIELD, THETA, n_steps, cutoff_frac=CUTOFF_FRAC, geometry=GEOMETRY,
        )

        if SIMULATE_MZ:
            # Simulate Mz via statevector on the logical circuit.
            # Transpilation preserves the unitary, so this equals running the transpiled circuit through a noiseless simulator.
            m.mz_sim = simulate_mz(circ, N_QUBITS)
        else:
            m.mz_sim = m.mz_exact

        all_metrics.append(m)

save_metrics(all_metrics, "../results/unitary_metrics.json")
print(f"Saved {len(all_metrics)} records")
```

## 3. Quick comparison with ideal

```python
ideal = load_metrics("../results/ideal_metrics.json")
ideal_1 = {m.alpha: m for m in ideal if m.n_steps == 1}
unitary_1 = {m.alpha: m for m in all_metrics if m.n_steps == 1}

print(f"{'alpha':>6}  {'pairs':>5}  {'ideal 2Q':>9}  {'uni 2Q':>7}  "
      f"{'tax':>5}  {'SWAPs':>6}")
print("-" * 48)
for alpha in ALPHAS:
    i, u = ideal_1[alpha], unitary_1[alpha]
    n_p = len(coupling_pairs(N_QUBITS, alpha, CUTOFF_FRAC))
    tax = u.two_q_depth / i.two_q_depth if i.two_q_depth > 0 else 0
    print(f"{alpha:>6.1f}  {n_p:>5}  {i.two_q_depth:>9}  {u.two_q_depth:>7}  "
          f"{tax:>5.2f}  {u.swap_count:>6}")
```

## Observations

- At $\alpha=4.0$ (NN only), the tax is **1.0** — no SWAPs needed.
- At $α\leq 1.0$ (all-to-all), the tax reaches ~1.8× in 2Q depth.
- SWAP count grows as more long-range pairs demand routing.
